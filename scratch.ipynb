```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

h = 6.626e-34
c = 2.9979e8
k = 1.3807e-23

b = 2.898e-3

T = 3.050e3

x = np.arange(1.0e-7, 1.0e-6, 0.1e-7)
y = (2. * h * c ** 2 / x ** 5) * (1. / (np.exp(h * c / (x * k * T)) - 1.))

fig = plt.figure(figsize = (20, 10))
ax1 = plt.subplot(111)
ax1.plot(x, y)
ax1.set_xlim(6e-7, 1e-6)
#ax1.set_ylim(2.7e12, 3.3e12)  # 3800.
ax1.set_ylim(0.55e12, 1.15e12)  # 3050.

ax1.plot([b / T, b / T], ax1.get_ylim(), 'k--')
```
